```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
file_path = 'GT_DataPD_MIT-CSPD.csv'
data = pd.read_csv(file_path)
data.head()
```

```python
x = data[['nqScore']].values 
y = data['gt'].values
x2 = data[['updrs108']].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
x2_train, x2_test, y_train, y_test = train_test_split(x2, y, test_size=0.2, random_state=0)
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
x2_train_scaled = scaler.fit_transform(x2_train)
x2_test_scaled = scaler.transform(x2_test)
```

```python
modelonqScore = LogisticRegression(random_state=0)
modeloUPDRS108 = LogisticRegression(random_state=0)
modelonqScore.fit(x_train_scaled, y_train)
modeloUPDRS108.fit(x2_train_scaled, y_train)
```

```python
y_pred = modelonqScore.predict(x_test_scaled)
y_pred2 = modeloUPDRS108.predict(x2_test_scaled)
```

```python
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
accuracy2 = accuracy_score(y_test, y_pred2)
report2 = classification_report(y_test, y_pred2)
```

```python
print(f"Precisión del modelo con nqScore: {accuracy * 100:.2f}%")
print("Reporte de clasificación con nqScore:")
print(report)
print(f"Precisión del modelo con UPDRS108: {accuracy2 * 100:.2f}%")
print("Reporte de clasificación con UPDRS108:")
print(report2)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
```

```python
tree_modelonqScore = DecisionTreeClassifier(random_state=0)
tree_modeloUPDRS108 = DecisionTreeClassifier(random_state=0)
tree_modelonqScore.fit(x_train_scaled, y_train)
tree_modeloUPDRS108.fit(x2_train_scaled, y_train)
```

```python
y_predTree = tree_modelonqScore.predict(x_test_scaled)
print("precision del modelo con nqScore(árbol de decisión): {:.2f}%".format(tree_modelonqScore.score(x_test_scaled, y_test)*100))   
print(classification_report(y_test, y_predTree))

y_predTree2 = tree_modeloUPDRS108.predict(x2_test_scaled)
print("precision del modelo con UPDRS108(árbol de decisión): {:.2f}%".format(tree_modeloUPDRS108.score(x2_test_scaled, y_test)*100))
print(classification_report(y_test, y_predTree2))

```

```python
from sklearn.svm import SVC
```

```python
modelo_svc_nq = SVC(kernel='rbf', random_state=0)
modelo_svc_nq.fit(x_train_scaled, y_train) 

modelo_svc_updrs = SVC(kernel='rbf', random_state=0)
modelo_svc_updrs.fit(x2_train_scaled, y_train)
```

```python
y_pred_svc_nq = modelo_svc_nq.predict(x_test_scaled)
print("precision del modelo con nqScore(SVM): {:.2f}%".format(modelo_svc_nq.score(x_test_scaled, y_test)*100))
print("reporte de clasificación con nqScore(SVM):")
print(classification_report(y_test, y_pred_svc_nq))

y_pred_svc_updrs = modelo_svc_updrs.predict(x2_test_scaled)
print("precision del modelo con UPDRS108(SVM): {:.2f}%".format(modelo_svc_updrs.score(x2_test_scaled, y_test)*100))
print("reporte de clasificación con UPDRS108(SVM):")
print(classification_report(y_test, y_pred_svc_updrs))
```
